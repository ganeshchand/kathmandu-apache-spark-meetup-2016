```
import sys
import time
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
#
#create streaming context with interval batches of one second.
#
ssc = StreamingContext(sc, 1)
```

Since I'm not using an external source to ingest data as streams, here we create RDD as batches and use the Queue as the input source. Good way to learn, try & do.
In this case create a queue of RDDs and push it onto the QueueInputDStream

```
rddQueue=[]
for i in range(5):
  rddQueue += [ssc.sparkContext.parallelize([j for j in range(1, 1001)], 10)]
  
```

```
# Create the QueueInputDStream and use it do some processing
inputStream = ssc.queueStream(rddQueue)
#map into pairs (v,v) where key is number divisible by 10
mappedStream = inputStream.map(lambda x: (x % 10, 1))
```

```
# reduce the stream by key and add the values
reducedStream = mappedStream.reduceByKey(lambda a, b: a + b)
```

```
#print the reduced stream
reducedStream.pprint()
```
